```python
from scipy.spatial import KDTree as kd_tree
import numpy as np
```

```python
from functional import seq
import csv
from tqdm import tqdm_notebook
from matplotlib.patches import Polygon
import matplotlib as mpl
import matplotlib.pyplot as plt

from ipywidgets import interact, interactive, fixed
import ipywidgets

 
%config InlineBackend.figure_format = 'svg' 

mpl.rcParams['figure.figsize'] = (9,8) 
```

```python
arr = np.random.uniform(-1, 1, (1000,2))
```

```python
tr = kd_tree(arr, 3)
```

```python
x = (0,0)
r = 0.5
inds = tr.query_ball_point(x, r)
fig, ax = plt.subplots()
ax.scatter(arr[:,0], arr[:,1])
ax.scatter(x[0], x[1])
ax.scatter(arr[inds,0], arr[inds,1])
ax.add_artist(plt.Circle(x,r,alpha=0.2))
plt.show()
```
