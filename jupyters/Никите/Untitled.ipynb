```python

bads = []
with open("powders_fortran.txt", 'w') as f:
    i = 1
    for p in bd:
        try:
            d = bd[p]
            print(f'case ({i}) ! {p}\n')
            print(f'   f_powder = {d["f"]*1e6:e}\n')
            print(f'   gamma = {1 + d["etta"]}\n')
            print(f'   covolume = {d["alpha_k"]*1e-3:e}\n')
            print(f'   T_1 = {d["T_1"]}\n')
            print(f'   powder_density = {d["ro"]*1000}\n')
            print(f'   I_k = {d["I_k"]*1e6:e}\n')
            print(f'   z_max = {d["Z_k"]}\n')
            print(f'   kappa_1 = {d["k_1"]}\n')
            print(f'   lambda_1 = {d["lambda_1"]}\n')
            print(f'   kappa_2  = {d["k_2"]}\n')
            print(f'   lambda_2 = {d["lambda_2"]}\n')
            print(f'   k_f = {d["k_f"]}\n')
            print(f'   k_I = {d["k_l"]}\n')
            print('\n')
            
            f.write(f'case ({i}) ! {p}\n')
            f.write(f'   f_powder = {d["f"]*1e6:e}\n')
            f.write(f'   gamma = {1 + d["etta"]}\n')
            f.write(f'   covolume = {d["alpha_k"]*1e-3:e}\n')
            f.write(f'   T_1 = {d["T_1"]}\n')
            f.write(f'   powder_density = {d["ro"]*1000}\n')
            f.write(f'   I_k = {d["I_k"]*1e6:e}\n')
            f.write(f'   z_max = {d["Z_k"]}\n')
            f.write(f'   kappa_1 = {d["k_1"]}\n')
            f.write(f'   lambda_1 = {d["lambda_1"]}\n')
            f.write(f'   kappa_2  = {d["k_2"]}\n')
            f.write(f'   lambda_2 = {d["lambda_2"]}\n')
            f.write(f'   k_f = {d["k_f"]}\n')
            f.write(f'   k_I = {d["k_l"]}\n')
            f.write('\n')
            i+=1
        except Exception as e:
            print('')
            print('==============================')
            print(bd[p], e)
            print('==============================')
            print('')
            bads.append(p)
```
