```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_theta_H2(T):
    return 0.3975 + 0.4814e-4 * T - 1.073e-7 * T**2 +4.6836e-11 * T**3 - 8.5361e-15 * T**4 + 5.6934e-19 * T**5

def get_theta_O2(T):
    return 0.4697 -2.9819e-4 * T +2.032e-7 * T**2 - 7.1938e-11 * T**3 + 12.311e-15 * T**4 -8.0736e-19 * T**5


def get_theta_H2O(T):
    return 0.3834 -1.8579e-4 * T +0.6266e-7 * T**2 - 1.0528e-11 * T**3 + 0.93554e-15 * T**4 -0.38512e-19 * T**5

def get_theta_He(T):
    return 0.67

Ts = np.linspace(50,5000, 10000)

plt.plot(Ts, get_theta_H2(Ts), label='$H_2$')
plt.plot(Ts, get_theta_O2(Ts), label='$0_2$')
plt.plot(Ts, get_theta_H2O(Ts), label='$H_2O$')

plt.grid()
plt.legend()
plt.show()
```
