```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import matplotlib as mpl
import numpy as np
import matplotlib.mlab as mlab
import math
%matplotlib inline 
%config InlineBackend.figure_format = 'svg' 

mpl.rcParams['figure.figsize'] = (6,3) 
# plt.xkcd(1,100,8)
```

```python
x = [0.5,2.5,5.5,8]
# plt.rcParams["font.family"] = "Times New Roman"
# plt.rcParams.update({'font.size': 22})
fig, ax = plt.subplots()

mu = 4
variance = 1
sigma = math.sqrt(variance)
xs = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
def func(x):
    return 1/sigma/math.sqrt(2*sigma*math.pi)*math.exp(-(x-mu)**2/2/sigma**2)
ys = np.asarray([func(x) for x in xs])
plt.plot(xs,ys,'black',linewidth=2)
plt.xticks(np.array(x))
plt.yticks([])
plt.ylim([0,1.1*max(ys)])
plt.xlim([0,1.1*max(x)])
plt.figtext(0.1, 0.9, '$n$')
plt.figtext(0.9, 0.05, 'iq?')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')

# We need to draw the canvas, otherwise the labels won't be positioned and 
# won't have values yet.
fig.canvas.draw()

labels = [item.get_text() for item in ax.get_xticklabels()]
labels[0] = ''
labels[1] = ''
labels[2] = ''
labels[3] = ''

def add_poly(a,b,col):
    ix = np.linspace(a, b)
    iy = np.asarray([func(x) for x in ix])
    verts = [(a, 0)] + list(zip(ix, iy)) + [(b, 0)]
    poly = Polygon(verts, color=col, alpha=0.5)
    ax.add_patch(poly)    

add_poly(1,2.5,'green')
add_poly(2.5,5.5,'blue')
add_poly(5.5,7,'orange')
ax.set_xticklabels(labels)

plt.savefig('11.png', bbox_inches='tight',dpi=300)
```

```python
x = [0.5,2,6]

fig, ax = plt.subplots()

mu = 2
variance = 1
sigma = math.sqrt(variance)
xs = np.linspace(0.5, mu + 3*sigma, 100)
def func(x):
    return mlab.normpdf(x, mu, sigma)
ys = func(xs)
plt.plot(xs,ys,'gray',linewidth=2)
plt.xticks(np.array(x))
plt.yticks([])
plt.ylim([0,1.1*max(ys)])
plt.xlim([0,1.1*max(x)])
plt.figtext(0.1, 0.9, 'p(x)')
plt.figtext(0.9, 0.05, 'x')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')

# We need to draw the canvas, otherwise the labels won't be positioned and 
# won't have values yet.
fig.canvas.draw()

labels = [item.get_text() for item in ax.get_xticklabels()]
labels[0] = 'a'
labels[1] = 'ген'
labels[2] = 'b'

def add_poly(a,b,col):
    ix = np.linspace(a, b)
    iy = func(ix)
    verts = [(a, 0)] + list(zip(ix, iy)) + [(b, 0)]
    poly = Polygon(verts, color=col, alpha=0.5)
    ax.add_patch(poly)    

# add_poly(1,2,'orange')
add_poly(0.5,5,'yellowgreen')
# add_poly(6,7,'orange')
ax.set_xticklabels(labels)

plt.savefig('2.png', bbox_inches='tight',dpi=300)
```
